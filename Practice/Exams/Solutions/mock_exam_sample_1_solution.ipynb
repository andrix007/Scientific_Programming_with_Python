```python
# Solutions 01

numbers = []  # Variables: to store user-entered numbers

while True:  # While loop: to repeatedly ask for numbers
    num = int(input("Enter a positive integer: "))  # Standard input: getting user input
    if num > 0:  # Relational operator: checking if the number is positive
        numbers.append(num)  # Variables: storing the number in a list
    else:
        print("Please enter a positive integer.")
        continue
    cont = input("Do you want to enter another number? (yes/no): ")  # Standard input
    if cont.lower() != 'yes':  # Logical operator and if statement: deciding to continue or not
        break  # Exiting the while loop

choice = input("Enter 'sum' to calculate sum, 'max' for maximum, 'min' for minimum: ")  # Standard input

match choice:  # Match statement: handling user's choice
    case 'sum':
        total = 0  # Variables: initializing sum
        for n in numbers:  # For loop: iterating over numbers
            total += n  # Arithmetic operator: summing numbers
        print(f"The sum is {total}")  # Standard output
    case 'max':
        maximum = numbers[0]  # Variables: initializing maximum
        for n in numbers:  # For loop: finding maximum
            if n > maximum:  # Relational operator: comparing numbers
                maximum = n
        print(f"The maximum number is {maximum}")  # Standard output
    case 'min':
        minimum = numbers[0]  # Variables: initializing minimum
        for n in numbers:  # For loop: finding minimum
            if n < minimum:  # Relational operator: comparing numbers
                minimum = n
        print(f"The minimum number is {minimum}")  # Standard output
    case _:  # Match default case
        print("Invalid command")  # Standard output
```

```python
# Solutions 03

from abc import ABC, abstractmethod

def create_shapes(*args, shape_type='Circle', **kwargs):  # Function with *args, **kwargs, default param
    shapes = []
    for dim in args:
        shape = globals()[shape_type](dim, **kwargs)  # Unpacking kwargs, keyword vs positional args
        shapes.append(shape)
    return shapes

def compute_total_area(shapes):  # Function definition
    total = 0
    for shape in shapes:
        total += shape.area()
    return total

class Shape(ABC):  # Abstract class, inheritance
    count = 0  # Class variable
    @abstractmethod
    def area(self): pass

class Circle(Shape):
    count = 0  # Class variable (name conflict resolution)
    def __init__(self, radius):
        self.radius = radius  # Instance variable
        Circle.count += 1
        Shape.count += 1
    def area(self):  # Method overriding
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        Shape.count += 1
    def area(self):
        return self.width * self.height

class Square(Rectangle):  # Inheritance
    def __init__(self, side):
        super().__init__(side, side)

class ColorMixin:
    def __init__(self, color='Red'):
        self.color = color

class ColoredSquare(Square, ColorMixin):  # Multiple inheritance
    def __init__(self, side, color='Red'):
        Square.__init__(self, side)
        ColorMixin.__init__(self, color)

shapes = create_shapes(2, 3, 4, shape_type='ColoredSquare', color='Blue')  # Function call
total_area = compute_total_area(shapes)  # Function call
print(f"Total area: {total_area}")
print(f"Shape count: {Shape.count}, Circle count: {Circle.count}")  # Class vs instance variable
for shape in shapes:
    print(f"Shape color: {shape.color}, Area: {shape.area()}")
```

```python
# Solutions 04

import numpy as np
import pandas as pd

data = np.genfromtxt('data.csv', delimiter=',')            # Load data from CSV (NumPy array creation)
col_means = np.nanmean(data, axis=0)                       # Aggregation function, operations on NaN
inds = np.where(np.isnan(data))                            # Find indices of NaNs
data[inds] = np.take(col_means, inds[1])                   # Replace NaNs with column means

mean = data.mean(axis=0)                                   # Aggregation function
std = data.std(axis=0)
normalized = (data - mean) / std                           # Vectorization and broadcasting

weights = np.linspace(1, 2, data.shape[1])                 # Create array using linspace
weighted = normalized * weights                            # Element-wise multiplication, upcasting

reshaped = weighted.reshape(-1, data.shape[1]*2)           # Array shape manipulation
split1, split2 = np.array_split(reshaped, 2)               # Splitting array into smaller ones

v_stacked = np.vstack((split1, split2))                    # NumPy.vstack()
h_stacked = np.hstack((split1, split2))                    # NumPy.hstack()

exp_array = np.exp(v_stacked)                              # Universal function

shallow_copy = exp_array.view()                            # Shallow copy (copy and views)
deep_copy = exp_array.copy()                               # Deep copy

dims, shape = exp_array.ndim, exp_array.shape              # Array attributes: ndim, shape
size, dtype = exp_array.size, exp_array.dtype
itemsize, data_buffer = exp_array.itemsize, exp_array.data

df = pd.DataFrame(exp_array)                               # Create DataFrame from NumPy array

df['Sum'] = df.sum(axis=1)                                 # Add new column (aggregation function)

df.iloc[[1, 3], [0, 2]] = np.nan                           # Random modification to cells (introduce NaNs)
df = df.dropna()                                           # Handle missing data (drop rows with NaN)

df['Category'] = pd.cut(df['Sum'], bins=3, labels=False)   # Convert numerical to categorical data

df = df[df['Category'] != 0]                               # Drop rows based on condition

print(df.head())                                           # Viewing data: head()
print(df.info())                                           # DataFrame info()

df.to_csv('modified_data.csv', index=False)                # Save DataFrame to CSV file
```

```python
# Solutions 05

import numpy as np
import scipy.integrate as integrate
import scipy.interpolate as interp
import scipy.ndimage as ndimage
import matplotlib.pyplot as plt

# Define quadratic function (SciPy integration of quadratic function)
def f(x, a=1, b=0, c=0):
    return a * x**2 + b * x + c

area, _ = integrate.quad(f, 0, 10)  # Integrate over interval [0, 10]

x = np.linspace(0, 10, 100)  # Generate data (NumPy array creation)
y = f(x)

y_smooth = ndimage.gaussian_filter(y, sigma=2)  # Apply Gaussian filter (signal processing)

interp_func = interp.interp1d(x, y, kind='cubic')  # Interpolation
x_new = np.linspace(0, 10, 400)
y_interp = interp_func(x_new)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))  # Create subplots (matplotlib.subplots())

# Plot original data (first subplot)
axs[0].plot(x, y, label='Original', color='blue', linestyle='-', linewidth=2)  # Styling line
axs[0].scatter(x, y, color='blue', marker='o', s=10)  # Markers
axs[0].set_title('Quadratic Function')  # Title
axs[0].set_xlabel('x Axis')  # x label
axs[0].set_ylabel('y Axis')  # y label
axs[0].legend()  # Legend
axs[0].grid(True)  # Grid
axs[0].tick_params(axis='both', which='both', direction='inout')  # Major and minor ticks

# Plot smoothed and interpolated data (second subplot)
axs[1].plot(x, y_smooth, label='Smoothed', color='red', linestyle='--')  # Smoothed data
axs[1].plot(x_new, y_interp, label='Interpolated', color='green', linestyle='-.')  # Interpolated data
axs[1].set_title('Smoothed and Interpolated Data')  # Title
axs[1].set_xlabel('x Axis')  # x label
axs[1].set_ylabel('y Axis')  # y label
axs[1].legend()  # Legend

ax_twin = axs[1].twinx()  # Additional axis object (twinx)
ax_twin.plot(x, y, label='Original Twin', color='gray', alpha=0.3)  # Plot on twinx
ax_twin.set_ylabel('Twin y Axis')  # y label for twinx

for ax in axs:
    ax.spines['top'].set_visible(False)  # Spine styling
    ax.tick_params(axis='x', which='major', labelsize=8)  # Major tick label size
#
plt.tight_layout()
plt.show()
```
